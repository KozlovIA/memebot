```python
from PIL import Image
import imagehash
import os

def get_hash(path):
    return imagehash.phash(Image.open(path))

folder = "memes_04.12.2025"
hashes = {}

for file in os.listdir(folder):
    path = os.path.join(folder, file)
    h = get_hash(path)
    hashes.setdefault(str(h), []).append(file)

# вывод похожих
for h, files in hashes.items():
    if len(files) > 1:
        print("Похожие файлы:", files)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Выводим похожие картинки рядом
for h, files in hashes.items():
    if len(files) > 1:
        print(f"Хэш: {h}")
        print("Похожие файлы:", files)

        n = len(files)
        plt.figure(figsize=(4 * n, 4))

        for i, fname in enumerate(files):
            img = Image.open(os.path.join(folder, fname))
            plt.subplot(1, n, i + 1)
            plt.imshow(img)
            plt.title(fname)
            plt.axis('off')

        plt.show()
```
